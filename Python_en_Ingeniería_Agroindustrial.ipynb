### 3 Ejemplos sencillos de Python en Agroindustria

Aquí tienes tres scripts básicos que muestran cómo Python puede facilitar tareas cotidianas:

#### 1. Cálculo del Índice de Cosecha

Este código ayuda a determinar qué tan eficiente fue una planta en convertir su biomasa en producto comercializable.

```
# Datos de entrada
peso_fruto = 450  # en gramos
biomasa_total = 1200  # en gramos (hojas, tallos, frutos)

# Cálculo del Índice de Cosecha (IC)
indice_cosecha = peso_fruto / biomasa_total

print(f"El Índice de Cosecha es: {indice_cosecha:.2f}")
# Un valor cercano a 1 indica alta eficiencia
```

#### 2. Monitoreo de Humedad con Alerta

Imagina que recibes datos de un sensor en un almacén de granos. Este script decide si hay que activar la ventilación.

```
humedad_actual = 18.5  # Porcentaje de humedad detectado

limite_seguro = 14.0

if humedad_actual > limite_seguro:
    print("ALERTA: Humedad alta detectada. Activando extractores.")
else:
    print("Estado: Humedad dentro de los rangos óptimos.")
```

#### 3. Conversión de Unidades (Rendimiento)

Es común necesitar pasar de toneladas por hectárea (t/ha) a kilogramos por metro cuadrado (kg/m²) para análisis más pequeños.

```
def rendimiento_a_kg_m2(ton_ha):
    # 1 ton = 1000 kg, 1 ha = 10000 m2
    return ton_ha * 0.1

rendimiento_maiz = 12.5  # t/ha
resultado = rendimiento_a_kg_m2(rendimiento_maiz)

print(f"El rendimiento es de {resultado} kg por cada metro cuadrado.")
```
